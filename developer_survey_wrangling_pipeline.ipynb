### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

#### 2. Explore the Dataset


<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
# Display column data types and non-null counts
df.info()
```

```python
# Show summary statistics and missing value information
df.describe(include='all')
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
# Show basic statistics for numerical columns only
df.describe()
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# View unique values in the Country column
df['Country'].value_counts().head(20)
```

```python
# Count missing Country values
df['Country'].isnull().sum()
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
# Standardize education levels into consistent categories
ed_map = {
    "Primary/elementary school": "Primary",
    "Secondary school (e.g. American high school, German Realschule or Gymnasium, etc.)": "Secondary",
    "Some college/university study without earning a degree": "Some College",
    "Associate degree (A.A., A.S., etc.)": "Associate",
    "Bachelor’s degree (B.A., B.S., B.Eng., etc.)": "Bachelors",
    "Master’s degree (M.A., M.S., M.Eng., MBA, etc.)": "Masters",
    "Professional degree (JD, MD, etc.)": "Professional",
    "Other doctoral degree (Ph.D., Ed.D., etc.)": "Doctorate"
}

df['EdLevel_Standardized'] = df['EdLevel'].map(ed_map)
df[['EdLevel','EdLevel_Standardized']].head()
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
# One-hot encode the Employment column
employment_encoded = pd.get_dummies(df['Employment'], prefix='Employment')

employment_encoded.head()
```

```python
# Add encoded columns to the dataset
df = pd.concat([df, employment_encoded], axis=1)
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
# Identify columns with the highest number of missing values
df.isnull().sum().sort_values(ascending=False).head(10)
```

```python
# Columns with extremely high missing values (>50% missing) provide limited analytical value
high_missing_cols = df.isnull().mean()
high_missing_cols = high_missing_cols[high_missing_cols > 0.5].index

# Drop highly incomplete columns
df = df.drop(columns=high_missing_cols)
df.shape
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
# Impute missing ConvertedCompYearly values using the median
median_salary = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(median_salary)

df['ConvertedCompYearly'].isnull().sum()
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
# Impute missing RemoteWork values using the most frequent category
df['RemoteWork'] = df['RemoteWork'].fillna(df['RemoteWork'].mode()[0])

df['RemoteWork'].isnull().sum()
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
# Apply Min-Max scaling to ConvertedCompYearly
min_val = df['ConvertedCompYearly'].min()
max_val = df['ConvertedCompYearly'].max()

df['ConvertedCompYearly_MinMax'] = (df['ConvertedCompYearly'] - min_val) / (max_val - min_val)

df[['ConvertedCompYearly', 'ConvertedCompYearly_MinMax']].head()
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
# Apply log transformation to reduce skewness
df['ConvertedCompYearly_Log'] = np.log10(df['ConvertedCompYearly'])

df[['ConvertedCompYearly', 'ConvertedCompYearly_Log']].head()
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
# Convert YearsCodePro to numeric
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Create ExperienceLevel categories
df['ExperienceLevel'] = pd.cut(
    df['YearsCodePro'],
    bins=[-1, 2, 5, 10, 20, 100],
    labels=['Junior', 'Mid', 'Senior', 'Lead', 'Expert']
)

df[['YearsCodePro', 'ExperienceLevel']].head()
```
